```python
from ThymioVision import ThymioVision 
import matplotlib.pyplot as plt
import Path_planning
import numpy as np
```

```python
frame = ThymioVision.captureImageOnCommand() #click c to snapshot after running
plt.imshow(frame)
```

```python
edges = ThymioVision.getEdges(frame, verbose=True)
```

```python
startPos = ThymioVision.detectBlueDot(frame, verbose=True)
print(startPos)
```

```python
goalPos = ThymioVision.detectGoal(frame, verbose=True)
print(goalPos)
```

```python
map = edges * (1/255)
plt.imshow(map)
plt.plot([startPos[0]], [startPos[1]], 'o')
plt.plot([goalPos[0]], [goalPos[1]], 'o')
```

```python
correctedStart = [startPos[1], startPos[0]]
correctedGoal = [goalPos[1], goalPos[0]]
path = Path_planning.A_star(map, correctedStart, correctedGoal)
```

```python
plt.imshow(frame)
for point in path:
    plt.plot(point[1], point[0], 'o')
```
